```python
from transformers import BertTokenizer, BertModel

# Load pre-trained model tokenizer (vocabulary) and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def text_to_vector(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().detach().numpy()

# Example: Vectorizing text data for one subcategory
vectorized_data = {}
for category, subcategories in data.items():
    vectorized_data[category] = {}
    for subcategory, texts in subcategories.items():
        vectorized_data[category][subcategory] = [text_to_vector(text) for text in texts]
```

```python
stop_words = set(stopwords.words('english'))

lemmatizer = WordNetLemmatizer()


processed_data = {
    'Finance': {},
    'Marketing': {},
    'HR': {}
}

for category, subcategories in data.items():
    for subcategory, docs in subcategories.items():
        docs_without_stopwords = []
        for doc in docs:
            words = doc.split()
            filtered_words = [word for word in words if word.lower() not in stop_words]
            filtered_doc = ' '.join(filtered_words)
            docs_without_stopwords.append(filtered_doc)
        processed_data[category][subcategory] = docs_without_stopwords
```

```python
topics_data = {
    'Finance': {},
    'Marketing': {},
    'HR': {}
}
topics_list=[]

def flatten(lst):
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)
    return flat_list


for category, subcategories in processed_data.items():
    for sub_category, docs in subcategories.items():
        model=BERTopic(language="english",nr_topics=5, min_topic_size=2)
        topics,probs=model.fit_transform(docs)
        topic_freq = model.get_topic_freq()
        topic_values = topic_freq['Topic'].tolist()
        for topic in topic_values:
            topics_tuples= model.get_topic(topic)
            topics_list.append([topic for topic, weight in topics_tuples])
        topics_data[category][sub_category] = list(set(flatten(topics_list)))

topics_data
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def normalize_vector(vector):
    norm = np.linalg.norm(vector)
    if norm == 0: 
       return vector
    return vector / norm

def retrieve_documents(prompt, vectorized_topics_data):
    prompt_vector = normalize_vector(get_sentence_embedding(prompt))
    best_match = None
    best_score = -1

    for category, subcategories in vectorized_topics_data.items():
        for subcategory, vector in subcategories.items():
                similarity = cosine_similarity([prompt_vector], [vector])
                if similarity > best_score:
                    best_score = similarity
                    best_match = (category, subcategory)
                # print((category, subcategory, best_score))
    
    return best_match

# prompt = "Looking for investment opportunities in the stock market"
prompt = "How many people are good at marketing in our company?"
# prompt = "Money reporting"

best_match = retrieve_documents(prompt, vectorized_topics_data)
print(f"Best Match: {best_match}")
```
